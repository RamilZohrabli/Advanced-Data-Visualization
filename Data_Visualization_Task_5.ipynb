```
import pandas as pd
import plotly.express as px

# Load the Iris dataset from seaborn (it can be used as an example)
from sklearn.datasets import load_iris

# Load the dataset
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)

# Add the species column to the dataframe
df['species'] = iris.target
df['species'] = df['species'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})

# Show the first few rows of the dataset
print(df.head())
```

```
# Create a simple scatter plot of sepal width vs sepal length
fig = px.scatter(df, x='sepal width (cm)', y='sepal length (cm)', color='species',
                 title="Sepal Width vs Sepal Length of Iris Flowers")

# Show the plot
fig.show()
```

```
# Create a scatter plot with hover data
fig = px.scatter(df, x='sepal width (cm)', y='sepal length (cm)', color='species',
                 title="Sepal Width vs Sepal Length",
                 hover_data=['species', 'sepal width (cm)', 'sepal length (cm)'])

# Show the plot
fig.show()
```

```
from plotly.subplots import make_subplots
import plotly.graph_objects as go

# Create subplots (1 row, 2 columns)
fig = make_subplots(rows=1, cols=2)

# Scatter plot in the first subplot
fig.add_trace(go.Scatter(x=df['sepal width (cm)'], y=df['sepal length (cm)'], mode='markers', name="Sepal Scatter"),
              row=1, col=1)

# Histogram in the second subplot
fig.add_trace(go.Histogram(x=df['sepal length (cm)'], name="Sepal Length Histogram"),
              row=1, col=2)

# Update the layout with a title
fig.update_layout(title_text="Multiple Visualizations")

# Show the plot
fig.show()
```

```
# Create a 3D scatter plot
fig = px.scatter_3d(df, x='sepal length (cm)', y='sepal width (cm)', z='petal length (cm)',
                    color='species', title="3D Scatter Plot of Sepal and Petal Length")

# Show the plot
fig.show()
```

```
# Create a scatter plot with slider functionality
fig = px.scatter(df, x="sepal width (cm)", y="sepal length (cm)", color="species",
                 title="Sepal Width vs Sepal Length with Filter")

# Update the layout with a slider to change data visibility
fig.update_layout(
    sliders=[{
        'steps': [{'label': 'Setosa', 'method': 'restyle', 'args': ['type', 'scatter']},
                  {'label': 'Versicolor', 'method': 'restyle', 'args': ['type', 'scatter']},
                  {'label': 'Virginica', 'method': 'restyle', 'args': ['type', 'scatter']}]
    }]
)

# Show the plot
fig.show()
```

```
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate correlation matrix
corr_matrix = df.iloc[:, :-1].corr()

# Create a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Heatmap of Iris Dataset")
plt.show()
```
